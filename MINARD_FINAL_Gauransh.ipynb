```python
import pandas as pd
import numpy as np
import plotly
plotly.offline.init_notebook_mode(connected=True)
import plotly.offline as py
import plotly.graph_objs as go
from plotly import tools

data = pd.read_csv("/users/gauranshraja/Desktop/minard-data.csv", header=0) 
data.head()



#Code providing the extension of line
sur1 = data[((data['DIR']=='A') & (data['DIV']==1)) | 
                   ((data['DIR']=='R') & (data['DIV']==1) & (data['SURV']==100000))]
sur2 = data[((data['DIR']=='A') & (data['DIV']==2)) | 
                   ((data['DIR']=='R') & (data['DIV']==2) & (data['LATP']==55.5))]
sur3 = data[(data['DIR']=='A') & (data['DIV']==3)]
sur4 = data[(data['DIR']=='R') & (data['DIV']==2)]
sur5 = data[(data['DIR']=='R') & (data['DIV']==1)]
sur6 = data[(data['DIR']=='R') & (data['DIV']==3) | 
                   (data['DIR']=='A') & (data['DIV']==3) & (data['LATP']==55.3)]

traceCities = go.Scatter(
    x = data.LONC,
    y = data.LATC,
    showlegend = False,
    mode ='markers+text',
    marker = dict(
        color = "black",
        size = 8
    ),
    text = data.CITY,
    textposition='middle right'
  )

trace1A = go.Scatter(
     x = sur1.LONP,
     y = sur1.LATP,
     name='ADVANCE',
    
     line = dict(
        color = ('#8c564b'),
        width = 10
        ),
  
     mode='lines'
)

trace1R = go.Scatter(
     x= sur5.LONP,
     y= sur5.LATP,
      name='RETREAT', 
  line = dict(
        color = ('#7e9cdf'),
        width = 8),
  mode='lines'
  )

#Code for extension of trace2A to trace2R
trace2A = go.Scatter(
     x= sur2.LONP,
     y= sur2.LATP,
     showlegend=False,

  line = dict(
        color = ('#8c564b'),
        width = 10),
  mode='lines',
  textposition='bottom center'

)

trace2R = go.Scatter(
#  sur = data[(data.SURV==340000)],
#     if (data.DIR=='A') and (data.DIV=='1'):
     x= sur4.LONP,
     y= sur4.LATP,
     showlegend=False,
    
  line = dict(
        color = ('#7e9cdf'),
        width = 8),
  mode='lines',
  text=sur4.CITY,
  )

#extension of 3A & 3R
trace3A = go.Scatter(   
    x= sur3.LONP,
    y= sur3.LATP,
    showlegend=False,
  
    line = dict(
        color = ('#8c564b'),
        width = 10),
        mode='lines',
        textposition='bottom center'
)

trace3R = go.Scatter(
    x = sur6.LONP,
    y = sur6.LATP,
    showlegend = False,
    
    line = dict(
        color = ('#7e9cdf'),
        width = 8
    ),
  
    mode = 'lines',
    textposition = 'bottom center'
)

traceTemp = go.Scatter(
    x = data.LONT,
    y = data.TEMP,
    showlegend = False,
    mode = 'lines+text+markers',
    marker = dict(
        symbol = 'square',
        color = "green",
        size = 8
    ),
    text = data.TEMP.fillna(0).astype(int).map(str) + ' On ' + data.MON + ' ' 
            + data.DAY.fillna(0).astype(int).astype(str),
    textposition = 'bottom center'
    
  )

# Temperature SUBPLOT
fig = tools.make_subplots(rows=2,cols=1)

fig.append_trace(trace1A,1,1)
fig.append_trace(trace1R,1,1)
fig.append_trace(trace2A,1,1)
fig.append_trace(trace2R,1,1)
fig.append_trace(trace3A,1,1)
fig.append_trace(trace3R,1,1)
fig.append_trace(traceCities,1,1)
fig.append_trace(traceTemp,2,1)


fig['layout'].update(
    title = dict(
        text = 'CHARLES JOSEPH MINARD’S MAP OF NAPOLEON’S RUSSIA CAMPAIGN.',
        font = dict(
            size = 20,
            family = "Droid Serif"
        ),
        xanchor = "center"
    ),
    
    margin = dict(
        b = 50,
        t = 50,
        pad = 5
    ),
    showlegend=True,
    width = 1000,
    height = 800,
    yaxis = dict(visible=False),
    yaxis2 = dict(title= 'Temperature',ticklen= 0.5,zeroline= False)
)

py.iplot(fig, validate = False)
```
